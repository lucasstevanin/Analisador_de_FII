```python
import requests
import pandas as pd
import numpy as np
```

```python
url = 'https://www.fundsexplorer.com.br/ranking'  
```

```python
response = requests.get(url)
if response.status_code == 200:
    df = pd.read_html(response.content, encoding='utf-8')[0]
```

```python
df.sort_values('Códigodo fundo', inplace=True)
```

```python
df.describe(include="all")
```

Dados Categoricos

```python
categorical_columns = ["Códigodo fundo", "Setor"]
```

```python
idx = df[df['Setor'].isna()].index
df.drop(idx, inplace=True)
```

```python
df[categorical_columns].isna().sum()
```

```python
df[categorical_columns] = df[categorical_columns].astype("category")
```

```python
df.info()
```

Dados numericos

```python
#  pega todas as colunas a partir da 2 coluna e menos a ultima coluna
col_floats = list(df.iloc[:,2:-1].columns)
```

```python
# substituir dados NaN por 0
df[col_floats] = df[col_floats].fillna(value=0)
```

```python
df[col_floats]
```

```python
df[col_floats].head()
```

```python
# Substituir simbolos por outros tudo de uma vez na tabela (poderia fazer atraves de codigos Regex)
df[col_floats] = df[col_floats].applymap(lambda x: str(x).replace('R$', '').replace('.0', '').replace('.', '').replace('%', '').replace(',', '.'))
```

```python
df[col_floats] = df[col_floats].astype('float')
```

```python
df.info()
```

```python
df.describe()
```

```python
# procurar por valores infinitos no dataframe
df[np.isinf(df[col_floats]).any(1)]

# Caso haja valor infinito... o codigo abaixo exclui o ativo com valor infinito
# idx = df[np.isinf(df[col_floats]).any(1)].index
# df.drop(idx, inplace=True)
```

```python
def oportunidade_media():
    
#     media_setor = df_aux.groupby('Setor').agg(['mean','std'])
    
#     df_setor = df[df[label_setor].isin([setor])]
#     df['Liquidez Diária'] = df['Liquidez Diária'] * df['Preço Atual']
    
    filter_ = \
        (df['QuantidadeAtivos'] > 6) &\
        (df['Liquidez Diária'] > 4500) &\
        (df['P/VPA'] < 1.1) &\
        (df['DY (12M)Acumulado'] > df['DY (12M)Acumulado'].mean())  
            
    print('média de Ativos: {}'.format(df['QuantidadeAtivos'].mean()))
    print('média de Liquidez Diária: {}'.format(df['Liquidez Diária'].mean()))
    print('média de P/VPA: {}'.format(df['P/VPA'].mean()))
    print('média de DY (12M) Acumulado: {}'.format(df['DY (12M)Acumulado'].mean()))

    
    return df_aux[filter_]
```

```python
oportunidade = oportunidade_media()
oportunidade.sort_values('Liquidez Diária', ascending=False, inplace=True)
oportunidade
```
